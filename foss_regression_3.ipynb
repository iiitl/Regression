# Feature Scaling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Reading the data for 3 columns
df = pd.read_csv('wines_SPA.csv', usecols=[3, 4, 7])
df.columns=['rating', 'number of reviews', 'price']

df
```

```python
#here, i have plotted the graph for the taken three columns
number_of_reviews = df['num_reviews']
price = df['price']
rating = df['rating']

plt.figure(figsize=(9, 4))
scatter = plt.scatter(rating, price, c=number_of_reviews, cmap='viridis')
plt.xlabel('rating')
plt.ylabel('Price')
plt.title('Wine data before feature scaling')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# implementing minmax scaling method of feature scaling for the given columns
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
number_of_reviews_scaled = scaler.fit_transform(number_of_reviews)
price_scaled = scaler.fit_transform(price)
rating_scaled = scaler.fit_transform(rating)

plt.figure(figsize=(9, 4))
plt.scatter(rating_scaled, price_scaled, c=number_of_reviews_scaled, cmap='viridis', label='Number of reviews')
plt.xlabel('rating')
plt.ylabel('price')
plt.title('Wine data after feature scaling(price, rating, number of reviews)')
plt.legend()
plt.grid(True)
plt.show()
```
