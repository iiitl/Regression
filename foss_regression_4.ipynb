# using numpy

```python
import pandas as pd
import numpy as np

data = pd.read_csv('wines_SPA.csv')

threshold = 4

#calculating z-scores for num_reviews, price, and acidity columns
z_score = np.abs((data[['num_reviews', 'price', 'acidity']] - data[['num_reviews', 'price', 'acidity']].mean()) / data[['num_reviews', 'price', 'acidity']].std())

new_data = data[(z_score < threshold).all(axis=1)]
print('after removing outliers')
print(new_data)
```

# visualising

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('wines_SPA.csv')

plt.figure(figsize=(10, 5))
plt.scatter(df.index, df['price'])
plt.xlabel('num_reviews')
plt.ylabel('price')
plt.title('Graph of wine prices with num_reviews')
plt.show()

#removing outliers by setting threshold value
threshold = 400
cleaned_df = df[df['price'] <= threshold]

plt.figure(figsize=(10, 5))
plt.scatter(cleaned_df.index, cleaned_df['price'])
plt.xlabel('num_reviews')
plt.ylabel('price')
plt.title('Wine prices (after removing outliers)')
plt.show()
```
