# Visualise the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#read the data
df = pd.read_csv('wines_SPA.csv')
```

```python
plt.figure(figsize=(19, 13))

plt.subplot(2, 3, 1)
sns.kdeplot(data=df, x='rating', hue='type', fill=True, common_norm=False)
plt.title('rating vs type')
```

```python
plt.figure(figsize=(10, 7))
plt.subplot(2, 3, 3)
sns.kdeplot(data=df, x='rating',y='price',fill=True, common_norm=False)
plt.title('Price vs rating')
```

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')

#converting 'wine' and 'region' columns to category codes
df['wine'] = df['wine'].astype('category').cat.codes
df['region'] = df['region'].astype('category').cat.codes

heatmap_cal = ['year','wine','region','rating','num_reviews','price','body','acidity']

matrix = df[heatmap_cal+['type']].corr()
```

```python
plt.figure(figsize=(9, 4))
sns.heatmap(matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Heatmap of the data')
plt.show()
```
